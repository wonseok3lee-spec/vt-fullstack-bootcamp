```python
import numpy as np

weather_data = np.array([
["2023-07-01", 30, 60, 0.1],
["2023-07-02", 31, 62, 0.2],
["2023-07-03", 29, 65, 0.0],
["2023-07-04", 33, 68, 0.5],
["2023-07-05", 34, 70, 0.3],
["2023-07-06", 32, 63, 0.1],
["2023-07-07", 30, 61, 0.0],
])
```

```python
# numpy type fix because everything is string now
dates = weather_data[:, 0]
temperature = weather_data[:, 1].astype(float)
humidity = weather_data[:, 2].astype(float)
precipitation = weather_data[:, 3].astype(float)
```

## Average temperature

```python
avg_temp = round(np.mean(temperature),2)
avg_temp
```

## Highest temp

```python
max_temp = np.max(temperature)
max_day = dates[np.argmax(temperature)]

max_temp, max_day
```

## Temperature daily fluctuations

```python
temp_diff = np.diff(temperature)
temp_diff
```

## high precipitation (> 0.2)

```python
high_precip_mask = precipitation > 0.2
high_precip_days = dates[high_precip_mask]

high_precip_days
```

## Average humidity on rainy days

```python
avg_humidity_rain = np.mean(humidity[high_precip_mask])
avg_humidity_rain
```

## Total precipitation

```python
total_precip = round(np.sum(precipitation),2)
total_precip
```

## Lowest temp

```python
min_temp = np.min(temperature)
min_day = dates[np.argmin(temperature)]

min_temp, min_day
```

## Summer days

```python
summer_mask = temperature > 30
summer_days = dates[summer_mask]

summer_days
```

## Average temp for summer days

```python
avg_summer_temp = np.mean(temperature[summer_mask])
avg_summer_temp
```

## No precipitation

```python
no_rain_days = precipitation == 0
count_no_rain = np.sum(no_rain_days)

count_no_rain
```

## Visualization

```python
import matplotlib.pyplot as plt

# Temp over time

plt.plot(dates, temperature, marker='o')
plt.title("Temperature Over Time")
plt.xticks(rotation=45)
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Relationship btw humidity vs rainfall

avg_humidity_rain = np.mean(humidity[precipitation > 0.2])
avg_humidity_no_rain = np.mean(humidity[precipitation <= 0.2])

labels = ['Rainy Days', 'Non-Rainy Days']
values = [avg_humidity_rain, avg_humidity_no_rain]

bars = plt.bar(labels, values)

# Add on top
for i, v in enumerate(values):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center')

plt.title("Average Humidity: Rainy vs Non-Rainy Days")
plt.ylabel("Humidity (%)")
plt.show()
```

## Conclusion

This analysis shows the weather patterns over the week.

The average temperature during the week was around 31°C. The highest temperature occurred on 2023-07-05, with the lowest 2023-07-03.

There were several days with high precipitation, especially on 2023-07-04 and 2023-07-05. 

The average humidity was higher on rainy days, indicating a relationship between rainfall and humidity.

The total precipitation for the week shows that there was moderate rainfall overall. There were also some dry days with no precipitation.

Summer days (temp > 30°C) occurred multiple times, and the average temperature on those days was higher than the weekly average.

Overall, the data provides useful insights into temperature trends, rainfall patterns, and humidity levels, which can help in understanding weather conditions.

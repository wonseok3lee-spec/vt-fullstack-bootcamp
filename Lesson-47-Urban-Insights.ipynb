## Lesson 47 Practice Project: Urban Insights – Data Analysis of Park Distribution in San Francisco

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, pearsonr, spearmanr

df = pd.read_csv('sf_parks.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['pop_category'] = pd.qcut(df['pop_density'], q=3, labels=['Low', 'Medium', 'High'])
df['pop_category'].head()
```

```python
df['pop_category'].value_counts()
```

```python
#turn it into category for chi-square test
df['income_category'] = pd.qcut(df['avg_income'], q=3, labels=['Low', 'Medium', 'High'])
df['income_category'].head()
```

```python
df['income_category'].value_counts()
```

```python
df.head()
```

```python
#turn it into category for chi-square test
df['parks_category'] = pd.qcut(df['num_parks'], q=3, labels=['Low', 'Medium', 'High'])
df['parks_category'].head()
```

```python
df['parks_category'].value_counts()
```

## Chi-square Test

```python
#contingency table: two categorical variables
#chi square test 1: population density and number of parks
contingency_pop = pd.crosstab(df['pop_category'], df['parks_category'])
contingency_pop
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_pop)

print("Chi2:", chi2)
print("p-value:", p)

#p-value = 0.6248 > 0.05: this indicates there's no statistically significant relationship between population density and number of parks
```

```python
#chi square test 2: average income and number of parks
contingency_income = pd.crosstab(df['income_category'], df['parks_category'])
contingency_income
```

```python
chi2_income, p_income, dof, expected = chi2_contingency(contingency_income)

print("Chi2:", chi2_income)
print("p-value:", p_income)

#p-value = 0.2014 > 0.05: this indicates there's no statistically significant relationship between average income and number of parks
```

```python
# neither population density nor average income appears to significantly influence park distribution
```

## Correlation Analysis

```python
#correlation analysis 1: population density and number of parks
corr_pop, p_pop_corr = pearsonr(df['num_parks'], df['pop_density'])

print("Correlation (parks vs pop_density):", corr_pop)
print("p-value:", p_pop_corr)

#There is no meaningful relationship between population density and number of parks.
```

```python
#correlation analysis 2: average income and number of parks
corr_income, p_income_corr = pearsonr(df['num_parks'], df['avg_income'])

print("Correlation (parks vs income):", corr_income)
print("p-value:", p_income_corr)

#There is a weak positive correlation between average income and number of parks, but it is not statistically significant.
```

## Final Verdict

Based on the analysis, park distribution does not show a statistically significant relationship with either population density or average income.

This suggests that current park placement may not be strategically aligned with community needs.

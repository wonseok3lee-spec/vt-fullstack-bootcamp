## Import the necessary libraries and create the dataset for analysis

```python
import pandas as pd
import numpy as np
from scipy.stats import skew

data = {
    'Product_A': [235, 189, 340, 270, 305, 210, 320, 265, 315, 280, 250, 200],
    'Product_B': [320, 190, 340, 270, 310, 260, 315, 290, 320, 275, 230, 320],
    'Product_C': [270, 290, 280, 240, 260, 210, 200, 340, 295, 280, 260, 310]
}

df = pd.DataFrame(data)
df
```

## Min-Max Normalization

```python
df_scaled = (df - df.min()) / (df.max() - df.min())

df_scaled.head()

# Min-Max scaling is applied to normalize the sales data between 0 and 1
```

## Basic Sales Analysis

```python
# Total Sales
total_sales = df.sum()
print("Total Sales:\n", total_sales)
```

```python
# Average Daily Sales
average_sales = df.mean()
print("Average Daily Sales:\n", average_sales)
```

```python
# Best-Selling Product
best_product = total_sales.idxmax()
best_value = total_sales.max()

print("Best Selling Product:", best_product)
print("Sales Value:", best_value)
```

```python
# Worst-Selling Product
worst_product = total_sales.idxmin()
worst_value = total_sales.min()

print("Worst Selling Product:", worst_product)
print("Sales Value:", worst_value)
```

```python
# Range of Sales
sales_range = df.max() - df.min()
print("Range of Sales:\n", sales_range)
```

## Advanced Statistical Analysis

```python
# Standard Deviation
std_dev = df.std()
print("Standard Deviation:\n", std_dev)

# Product A has the highest variability, while Product C is the most stable
```

```python
# Skewness
skewness = df.apply(skew)
print("Skewness:\n", skewness)

# All products show slight negative skewness, meaning sales tend to have a longer tail on the lower side
```

```python
# Coefficient of Variation
cv = df.std() / df.mean()
print("Coefficient of Variation:\n", cv)

# Product A has the highest relative variability, while Product C is the most consistent
```

```python
# Product A shows the highest variability, while Product C is the most stable across all metrics
```

```python
# Percentiles
```

```python
# 25th Percentile: at lower 25% point
percentile_25 = df.quantile(0.25)
print("25th Percentile:\n", percentile_25)
```

```python
# 75th Percentile: at higher 25% point
percentile_75 = df.quantile(0.75)
print("75th Percentile:\n", percentile_75)
```

```python
# IQR 
iqr = percentile_75 - percentile_25
print("Interquartile Range (IQR):\n", iqr)
```

## Final Summary

Based on the basic/advanced statistical analysis of the retail sales data, several key insights were identified.

Product B recorded the highest total sales, making it the best-performing product overall, while Product A had the lowest total sales. In terms of variability, Product A showed the highest standard deviation and coefficient of variation, indicating that its sales are more inconsistent. On the other hand, Product C demonstrated the lowest variability, making it the most stable product.

The skewness values for all products were slightly negative, suggesting that the sales distributions are slightly left-skewed, with occasional lower sales values.

Furthermore, the interquartile range (IQR) analysis showed that Product A has the widest spread in the middle 50% of its data, while Product C has the most concentrated distribution.

Overall, Product B performs best in terms of sales volume, Product C is the most stable, and Product A shows the highest variability. These insights can help the company make informed decisions regarding inventory management, marketing strategies, and product focus.
